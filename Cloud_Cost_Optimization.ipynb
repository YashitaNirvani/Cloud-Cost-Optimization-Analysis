```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/gcp_final_approved_dataset.csv')
```

Data Cleaning

```python
df["Usage Start Date"] = pd.to_datetime(df["Usage Start Date"], format='%d-%m-%Y %H:%M')
df["Usage End Date"] = pd.to_datetime(df["Usage End Date"], format='%d-%m-%Y %H:%M')
```

```python
df["UsageDuration(Hours)"] = (df["Usage End Date"] - df["Usage Start Date"]).dt.total_seconds()/3600
```

Which services cost the most?

```python
service_cost = (
    df.groupby("Service Name")["Total Cost (INR)"]
    .sum()
    .sort_values(ascending=False)
)

print(service_cost)
```

```python
service_cost_usd = (
    df.groupby("Service Name")["Rounded Cost ($)"]
    .sum()
    .sort_values(ascending=False)
)

print(service_cost_usd)
```

```python
service_cost.plot(kind="bar", figsize=(10,5))
plt.title("Total Cost by Cloud Service (INR)")
plt.ylabel("Cost (INR)")
plt.xlabel("Service Name")
plt.show()
```

Are resources underutilized but expensive?

```python
avg_cost = df["Total Cost (INR)"].mean()
```

```python
underutilized = df[
    (df["CPU Utilization (%)"] < 40) &
    (df["Memory Utilization (%)"] < 50) &
    (df["Total Cost (INR)"] < avg_cost)
    .sort_values(ascending=False)
]
```

```python
underutilized[[
    "Service Name",
    "CPU Utilization (%)",
    "Memory Utilization (%)",
    "Total Cost (INR)"
]]
```

Region-wise Cost Analysis

```python
region_cost =(
    df.groupby("Region/Zone")["Total Cost (INR)"]
    .sum()
    .sort_values(ascending=False)
)

print(region_cost)
```

```python
region_cost.plot(kind="pie", autopct="%1.1f%%", figsize=(7,7))
plt.title("Cost Distribution by Region")
plt.ylabel("")
plt.show()
```

Network Usage vs Cost

```python
df["Total Network (GB)"]=(
    df["Network Inbound Data (Bytes)"]+
    df["Network Outbound Data (Bytes)"]
) / (1024**3)

```

```python
sns.scatterplot(
    data=df,
    x="Total Network (GB)",
    y="Total Cost (INR)",
    hue="Service Name"
)
plt.title("Network Usage vs Cost")
plt.show()
```
